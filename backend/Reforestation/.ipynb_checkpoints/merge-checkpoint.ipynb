```python
import pandas as pd
```

```python
filter_df = pd.read_csv('filter-reforest.csv')
feature_df = pd.read_csv('feature_classes.csv')
```

```python
# Drop overlapping columns in df2 (excluding merge keys)
cols_to_drop = [col for col in filter_df.columns if col in feature_df.columns and col not in ['State', 'Abbreviation', 'County']]
df2_revised = filter_df.drop(cols_to_drop, axis=1)


df_merged = pd.merge(feature_df, df2_revised, on=['State', 'Abbreviation', 'County'], how='left')

df_merged.fillna(0, inplace=True)

df_merged.head()

```

```python
df_merged.to_csv('reforestation.csv', index=False)
```

```python
#only michigan (initial code)
michigan_df = df_merged[df_merged['State'] == 'Michigan']
michigan_df.head()
```
