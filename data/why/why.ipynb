# Why is carbon removal important?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('carbon.csv')
df.head()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df['date'] = pd.to_datetime(df[['Year', 'Month']].assign(DAY=1))

sns.set_style("whitegrid")

sns.lineplot(x='date', y='Seasonally Adjusted CO2 Fit (ppm)', data=df, color="b", lw=2)

plt.title('Trend of Carbon Dioxide (CO2) Levels Over Time')
plt.xlabel('Year')
plt.ylabel('Carbon Dioxide (ppm)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
